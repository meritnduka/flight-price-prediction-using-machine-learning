# Feature Engineering:  process of creating a new column from a manipulated column by use of domain knowledge

```python
df["month_of_Journey"]=df["Date_of_Journey"].dt.month
df["day_of_Journey"]=df["Date_of_Journey"].dt.day_name()
df["year_of_Journey"]=df["Date_of_Journey"].dt.year
```

```python
df.drop(columns=["Date_of_Journey","Route","Dep_Time","Arrival_Time"], inplace=True)
```

```python
df["Duration"].value_counts()
```

```python
def extract(x):
    hour=0
    minute=0
    x= x.split()
    if len(x)==1:
        if x[0].endswith("h"):
            hour = int(x[0].strip("h"))
        else:
            minute= int(x[0].strip("m"))
    else:
        hour = int(x[0].strip("h"))
        minute = int(x[1].strip("m"))
        
        
        
    return hour, minute
```

```python
hour= df["Duration"].apply(lambda p: extract(p)[0])*60
minute= df["Duration"].apply(lambda p: extract(p)[1])
df["duration in minute"]= hour+minute
```

```python
df.drop(columns=["Duration"], inplace=True)
```

```python
df["Airline"].replace(["Jet Airways Business", "Multiple carriers Premium economy", "Vistara Premium economy"],
                     ["Jet Airways", "Multiple carriers", "Vistara"], inplace=True)
```

```python
df["Airline"].value_counts()
```

```python
df_d= pd.get_dummies(df, columns=["Airline","Source", "Destination", "Total_Stops", "Additional_Info"], drop_first=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
df_d["day_of_Journey"]=le.fit_transform(df_d["day_of_Journey"])
```
